```python
import os
from dualneuron.dream.sim import embeddings
import matplotlib.pyplot as plt
import numpy as np

from dotenv import load_dotenv
load_dotenv();

token = os.getenv("HF_TOKEN")
data_dir = os.getenv("DATA_DIR")
```

```python
out = embeddings(
    data_dir=data_dir + "datasets", 
    cache_dir=data_dir + "dreamsim_models",
    token=token, 
    split='test', 
    dataset="imagenet",
    area='v4',
    use_grayscale=False,
    use_mask=True,
    use_norm=True,
    norm=80.0,
    num_channels=3,
    crop_size=236,
    bg_value=0.0,
    batch_size=32,
    num_workers=0,
)
```

```python
norms = []
for i in range(out.shape[0]):
    img_norm = np.linalg.norm(out[i])
    norms.append(img_norm)

norms = np.array(norms)

print(f"Number of images: {len(norms)}")
print(f"Mean norm: {norms.mean():.4f}")
print(f"Median norm: {np.median(norms):.4f}")
print(f"Std norm: {norms.std():.4f}")
print(f"Min norm: {norms.min():.4f}")
print(f"Max norm: {norms.max():.4f}")
```
